In trigram representation, it could be "He likes cat". Explain why the storage and computation cost increase with the growth of n in n-gram methods. 

Answer to Q1: 

There is a possibility that there are n^2 separate bigrams, each of which cannot follow the previous word, for every n distinct words. In addition, there are x various x grams in relation to the quantity of words that are offered. Adding a note that the gram bag is significantly bigger and has less of it. All of the aforementioned justifies why n grams are more expensive to compute and store. As a result, as data volume grows, so do the costs of computation and storage.

## (Tutorial) Stemming and Lemmatization

```python
# import PorterStemmer class form nltk.stem.porter module
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer() 

stem = stemmer.stem('flowers')  
print(f"'flowers' after stemming: {stem}")

stem = stemmer.stem('zeroes')
print(f"'zeroes' after stemming: {stem}")

stem = stemmer.stem('better')
print(f"'better' after stemming: {stem}")

stem = stemmer.stem('sixties')
print(f"'sixties' after stemming: {stem}")

stem = stemmer.stem('goes')
print(f"'goes' after stemming: {stem}")

stem = stemmer.stem('go')
print(f"'go' after stemming: {stem}")
```

**Answer to Q2**:

Stemming is a rule-based method. Lematizer is a method based on canonical dictionaries. When comparing lemmatization and stemming in terms of accuracy and performance, lemmatization always wins. Lematization lends itself well to context-based approaches. Stemming, on the other hand, is not suitable for contextual analysis. Good if context is not the primary consideration. Compared to stemming, lematization takes longer. The prefixes and suffixes of the last word are removed during stemming, thus losing their meaning. Rematization does not. ### Answer to Q4: 

Bag of words disregards the word's semantics.
The most pressing issue in the bag of words is also the variety of vocabulary.
If your corpus is large, with many statements, the resulting matrix is sparse.

```python
import pandas as pd
import numpy as np
import collections
```

```python
Sen1="The Bahubali movie is the extraordinary movie in the world!"
Sen2="The Bahubali movie is the best movie!"
Sen3="The Bahubali is nice"
```

```python
import re
```

```python

Sen1= re.sub(r"[^a-zA-Z0-9]", " ", Sen1.lower()).split()
Sen2= re.sub(r"[^a-zA-Z0-9]", " ", Sen2.lower()).split()
Sen3= re.sub(r"[^a-zA-Z0-9]", " ", Sen3.lower()).split()
```

```python
Sen1
```

```python
wordnet_12= np.union1d(Sen1,Sen2)
wordnet =  np.union1d(wordnet_12,Sen3)
print(wordnet)
```

```python
def Bag_of_words(wordnet,Sen):
  dis = dict.fromkeys(wordnet,0)
  for i in Sen:
     dis[i]=Sen.count(i)
  return dis
```

```python
bow1 = Bag_of_words(wordnet,Sen1)
bow2 = Bag_of_words(wordnet,Sen2)
bow3 = Bag_of_words(wordnet,Sen3)
df_bow = pd.DataFrame([bow1,bow2,bow3])
df_bow.head()
```
